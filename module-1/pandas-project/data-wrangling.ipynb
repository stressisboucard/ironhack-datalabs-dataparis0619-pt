```python
def getmonth(x):
    month = re.findall('\-\w{3}\-', x)
    if len(month) == 0:
        return None
    return month[0].strip('-')

def getdays(x):
    days = re.findall('\-\d{2}',x)
    if len(days)==0: 
        return None
    return days[0].strip('-')
```

```python
def clean(name):
    re_name = '(([A-Z]{1}[a-z]+\s([a-z]+\s)?)?[A-Z]{1}[a-z]+)'
    clean_name = ''
    if name[0].isupper() and name[0].isalpha():
        matches = re.findall(re_name,name)
        if matches: 
            detected_names = [match[0] for match in matches]
            return ', '.join(detected_names)
    else: return None

Victim = s_attack1[['Victim']].applymap(clean)
```
